# Load Data

```python
from tabpfn_project.paths import RESULTS_DIR
from tabpfn_project.helper.utils import load_pickle

tabpfn_results_dir = RESULTS_DIR / "tabpfn"
distnet_results_dir = RESULTS_DIR / "distnet"
rf_results_dir = RESULTS_DIR / "random_forest"
gp_results_dir = RESULTS_DIR / "gaussian_process"
bayesian_distnet_results_dir = RESULTS_DIR / "bayesian_distnet"

tabpfn_scalerLOG = load_pickle(tabpfn_results_dir / "tabpfn_scalerLOG.pkl")
tabpfn_scalerMAX = load_pickle(tabpfn_results_dir / "tabpfn_scalerMAX.pkl")
tabpfn_scalerLOG_noDUPS = load_pickle(
    tabpfn_results_dir / "tabpfn_scalerLOG_noDUPS.pkl"
)
tabpfn_naive = load_pickle(tabpfn_results_dir / "tabpfn_naive.pkl")
tabpfn_oracle = load_pickle(tabpfn_results_dir / "tabpfn_oracle.pkl")

distnet_scalerLOG = load_pickle(distnet_results_dir / "distnet_scalerLOG.pkl")
distnet_scalerMAX = load_pickle(distnet_results_dir / "distnet_scalerMAX.pkl")

bayesian_distnet_scalerMAX = load_pickle(
    bayesian_distnet_results_dir / "bayesian_distnet_scalerMAX.pkl"
)

random_forest_newDEFS = load_pickle(rf_results_dir / "random_forest_newDEFS.pkl")
random_forest_oldDEFS = load_pickle(rf_results_dir / "random_forest_oldDEFS.pkl")
random_forest_tuned = load_pickle(rf_results_dir / "random_forest_tuned.pkl")

gp_results = load_pickle(gp_results_dir / "gp_scalerLOG.pkl")
```

# Load Plotting Function

```python
from tabpfn_project.helper.utils import plot_main_results
```

# Ablation 1: Log-scaler vs. Min-Max scaler for TabPFN

```python
results_dict = {
    "TabPFN (scaler=LOG)": tabpfn_scalerLOG,
    "TabPFN (scaler=MAX)": tabpfn_scalerMAX,
}

visual_cfg = {
    "TabPFN (scaler=LOG)": ("#2CA02C", "xx"),
    "TabPFN (scaler=MAX)": ("#D62728", "--"),
}

fig = plot_main_results(
    results_dict,
    visual_cfg,
    log_x=True,
    log_y=True,
    zoom_on_model="TabPFN (scaler=LOG)",
)
fig.savefig(fname="tabpfn_scaler_effect.pdf")
```

# Ablation 2: With/Without repetitions' effect on performance of TabPFN

```python
results_dict = {
    "TabPFN": tabpfn_scalerLOG,
    "TabPFN (w/o duplicate instances)": tabpfn_scalerLOG_noDUPS,
}

visual_cfg = {
    "TabPFN": ("#2CA02C", "xx"),
    "TabPFN (w/o duplicate instances)": ("#D62728", "--"),
}

fig1 = plot_main_results(
    results_dict,
    visual_cfg,
    log_x=True,
    log_y=True,
    zoom_on_model="TabPFN (w/o duplicate instances)",
)
fig1.savefig("tabpfn_noDUPS_all.pdf")
```

# Ablation 3: Log-scaler vs. Min-Max scaler for DistNet

```python
results_dict = {
    "DistNet (scaler=LOG)": distnet_scalerLOG,
    "DistNet (scaler=MAX)": distnet_scalerMAX,
}

visual_cfg = {
    "DistNet (scaler=LOG)": ("#1F77B4", "//"),
    "DistNet (scaler=MAX)": ("#D62728", "--"),
}

fig = plot_main_results(
    results_dict,
    visual_cfg,
    log_x=True,
    log_y=True,
    zoom_on_model="DistNet (scaler=MAX)",
)
fig.savefig("distnet_scaler_effect.pdf")
```

# Ablation 4: Effect of new hyperparameters on Random Forest

```python
results_dict = {
    "Random Forest (new defaults)": random_forest_newDEFS,
    "Random Forest (old defaults)": random_forest_oldDEFS,
}

visual_cfg = {
    "Random Forest (new defaults)": ("#FF7F0E", "\\"),
    "Random Forest (old defaults)": ("#D62728", "oo"),
}


fig1 = plot_main_results(results_dict, visual_cfg, log_x=True, log_y=True)
fig1.savefig("random_forest_comparisons_newVsOld_defaults.pdf")
```

# Ablation 5: Effect of hyperparameter tuning on Random Forest

```python
results_dict = {
    "Random Forest": random_forest_newDEFS,
    "Random Forest (tuned)": random_forest_tuned,
}

visual_cfg = {
    "Random Forest": ("#FF7F0E", "\\"),
    "Random Forest (tuned)": ("#1DF9F9", "oo"),
}


fig1 = plot_main_results(results_dict, visual_cfg, log_x=True, log_y=True)
fig1.savefig("random_forest_comparisons_tuned_vs_newDefaults.pdf")
```
