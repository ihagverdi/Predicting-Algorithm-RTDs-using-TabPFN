# Load Data

```python
from tabpfn_project.paths import RESULTS_DIR
from tabpfn_project.helper.utils import load_pickle

tabpfn_results_dir = RESULTS_DIR / "tabpfn"
distnet_results_dir = RESULTS_DIR / "distnet"
rf_results_dir = RESULTS_DIR / "random_forest"
gp_results_dir = RESULTS_DIR / "gaussian_process"

tabpfn_scalerLOG = load_pickle(tabpfn_results_dir / "tabpfn_fdrop.pkl")
distnet_scalerMAX = load_pickle(distnet_results_dir / "distnet_fdrop.pkl")
random_forest_newDEFS = load_pickle(rf_results_dir / "random_forest_fdrop.pkl")
gp_results = load_pickle(gp_results_dir / "gp_fdrop.pkl")
```
